#  **Task 1 - Problem Statement**
 
Cyclistic, a bike-sharing company, believes the future of the company depends on maximizing the number of annual membership. It currently has two categories of users; casual riders (pay-as-you-go) and annual members (yearly subscribers). This case study seeks to understand how casual riders and annual members use the bikes differently using the company's data over a 12 months period (Mar. 2021 - Apr. 2022). Based on the insights gained, recommendations will be made on new marketing strategy to convert casual riders into annual members. The data is accessible [here](https://divvy-tripdata.s3.amazonaws.com/index.html) through Motivate International Inc. under this [license](https://ride.divvybikes.com/data-license-agreement).

**Step 4 - Combine All Datasets into a Single Dataframe**

# **Task 3 - Process Data from Dirty to Clean**

**Step 1 - Inspect the New Table Created**

```r
colnames(all_trips)  
nrow(all_trips) 
dim(all_trips)
head(all_trips)  
str(all_trips)  
summary(all_trips)  
```

**Step 2 - Add Columns that List the Date, Month, day, and Year of Each Ride**

```r
all_trips$date <- as.Date(all_trips$started_at) 
all_trips$month <- format(as.Date(all_trips$date), "%m")
all_trips$day <- format(as.Date(all_trips$date), "%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")
```

**Step 3 - Add a column for the Ride Duration of Each Ride**

```r
all_trips$ride_length <- (as.double(difftime(all_trips$ended_at,all_trips$started_at)))/60
all_trips$time <- format(all_trips$started_at, format = "%H:%M")
all_trips$time <- as.POSIXct(all_trips$time, format = "%H:%M")
```

**Step 4 - Inspect the Structure of the columns**

```r
str(all_trips)
```

**Step 5 - Convert New "ride_length" Column From Factor to Numeric**

```r
is.factor(all_trips$ride_length)
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))
is.numeric(all_trips$ride_length)
```

**Step 6 - Remove Bad Data (negative trip) & Create a New Version of Dataframe**

```r
all_trips_v2 <- all_trips[!(all_trips$ride_length<0),]
```

# **Task 4 - Analyze Data**

**Step 1 - Perform Descriptive Analysis on Ride Length**

```r
summary(all_trips_v2$ride_length)
```

**Step 2 - Compare the Trip Details of Members and Casual User**

```r
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = mean)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = median)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = max)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = min)
```

**Step 3 - Compute Average Ride Time by Each Day for Members v. Casual Users**

```r
all_trips_v2$day_of_week <- ordered(all_trips_v2$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
```

**Step 4 - Analyze Ridership Data by Type & Weekday**

```r
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>%  
  group_by(member_casual, weekday) %>%  
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) 
```

**Step 5 - Analyze Ridership Data by Type & Month**

```r
all_trips_v2 %>% 
  mutate(month = month(started_at, label = TRUE)) %>%  
  group_by(member_casual, month) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, month)
```

# **Task 5 - Visualize Data**

**Step 1 - Visualize Number of Rides by User Type and Weekday**

```r
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = number_of_rides, fill = member_casual)) +
  geom_col(position = "dodge")+ labs(title = "NUMBER OF RIDES BY USERS FOR EACH DAY OF THE WEEK", subtitle = "Rides between Apr 2021 - Mar 2022")
```

***Insights***

The chart above shows that casual riders use bikes more on the weekends (Saturday and Sunday) than weekdays. Whereas, the demand for bikes on the weekdays is dominated by annual members. We can also notice that there is less spread in bike demand for annual members throughout the week as compared to the huge variation in casual member's demand between weekdays and weekends.

**Step 2 - Visualize Number of Rides by User Type and Month**

```r
all_trips_v2 %>% 
  mutate(month = month(started_at, label = TRUE)) %>% 
  group_by(member_casual, month) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, month)  %>% 
  ggplot(aes(x = month, y = number_of_rides, fill = member_casual)) +
  geom_col(position = "dodge")+ labs(title = "NUMBER OF RIDES BY USERS PER MONTH", subtitle = "Rides between Apr 2021 - Mar 2022")
```

***Insights***

The chart above shows that summer months (June, July, & August) are the peak periods and there is more demand from casual riders than annual members. This might be as a result of the favorable summer weather, which attracts tourists demand for the bikes.  However, annual members demand were greater for every other months of the year especially in the winter months ( December, January, & February) where we notice huge disparities. 

**Step 3 - Visualize Average Ride Duration by User Type & Weekday**

```r
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = average_duration, fill = member_casual)) +
  geom_col(position = "dodge") + labs(title = "AVERAGE RIDE DURATION EACH WEEK DAY", subtitle = "Rides between Apr 2021 - Mar 2022")
```

***Insights***

The data above shows that the average ride duration of casual riders is at least twice that of the members every single day of the week. The chart also indicates that casual riders' average trip during the weekend (Saturday & Sunday) is distintively greater than weekday trips just like the "Number of Rides" data shows. 


**Step 4 - Visualize Average Ride Duration by User Type & Month**

```r
all_trips_v2 %>% 
  mutate(month = month(started_at, label = TRUE)) %>% 
  group_by(member_casual, month) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, month)  %>% 
  ggplot(aes(x = month, y = average_duration, fill = member_casual)) +
  geom_col(position = "dodge") + labs(title = "AVERAGE RIDE DURATION FOR EACH MONTH", subtitle = "Rides between Apr 2021 - Mar 2022")
```

***Insights***

This data shows that casual riders' average trip duration is higher than that of the members for every month of the year. While we see a steady pattern in members'trip duration ranging from 12-15 minutes throughout the year, we notice a huge disparity in casual riders' trip duration from 23-38 minutes. Summer months again prove to have peak average trip duration.

**Step 5 - Visualize Average Bike Demand over the Hours of the Day**

```r
all_trips_v2 %>%  
  group_by(member_casual, time) %>% 
  summarise(number_of_trips = n()) %>%
  ggplot(aes(x = time, y = number_of_trips, color = member_casual, group = member_casual)) +
  geom_line() +
  scale_x_datetime(date_breaks = "1 hour", minor_breaks = NULL,
                   date_labels = "%H:%M", expand = c(0,0)) +
  theme(axis.text.x = element_text(angle = 90)) +
  labs(title ="Demand over 24 hours of a day", x = "Time of the day")
```

***Insights***

The annual members demands for bikes seem to peak around 8am and 5pm. These coincide with conventional start time and end time of typical office work respectively. The demand for casual riders peaks once around 5pm. Generally, we can observe that annual members demand for bike exceeds that of casual members during the day time while that of casual members is slightly higher during the night between 9pm-5am.
